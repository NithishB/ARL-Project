```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Perceptron
from sklearn.pipeline import Pipeline
```

```python
points = []
cap = cv2.VideoCapture('top.mp4')
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret:
        img_grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        ret,img_grey = cv2.threshold(img_grey,5,255,cv2.THRESH_BINARY)
        contours, hierarchy = cv2.findContours(img_grey,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
        plain_img = np.ones_like(img_grey)*255
        selected_contours = []
        for cnt in contours:
            if cv2.contourArea(cnt) < 3:
                selected_contours.append(cnt)
        points.append(sorted(np.array(selected_contours).ravel().reshape((-1,2)), key=lambda x: x[0]))
        plain_img = cv2.drawContours(plain_img, selected_contours, -1, (0,255,0), 3)
        cv2.imshow('Frame',plain_img)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else: 
        break
cap.release()
cv2.destroyAllWindows()
```

```python
all_x, all_y = [], []
collect_x, collect_y = [], []
for i,ps in enumerate(points):
    ps = np.array(ps)
    indices = np.where((ps[:,0]>=150)&(ps[:,0]<=800))
    x,y = ps[indices,0], ps[indices,1]
    collect_x.append(x.tolist()[0])
    collect_y.append(y.tolist()[0])
    all_x.extend(x.tolist()[0])
    all_y.extend(y.tolist()[0])
```

```python
new_x, new_y = [], []
for i in range(len(all_x)):
    if all_x[i] not in new_x:
        new_x.append(all_x[i])
        new_y.append(all_y[i])
```

```python
model = Pipeline([('poly', PolynomialFeatures(degree=5)),
                  ('linear', LinearRegression(fit_intercept=False))])

for i in range(len(collect_x)):
    x,y = np.array(collect_x[i]), np.array(collect_y[i])
    model = model.fit(x[:, np.newaxis], y)
    xn = np.arange(150,800,3)
    yn = model.predict(xn[:, np.newaxis])
    plt.plot(xn,yn)
#     print(model.named_steps['linear'].coef_)
plt.show()
```

```python
model = Pipeline([('poly', PolynomialFeatures(degree=5)),
                  ('fitter', LinearRegression(fit_intercept=True))])
x,y = np.array(new_x), np.array(new_y)
split=int(0.7*len(x))
x_train, x_test = x[:split], x[split:]
y_train, y_test = y[:split], y[split:]

model = model.fit(x_train[:, np.newaxis], y_train)
y_pred = model.predict(x_test[:, np.newaxis])
plt.scatter(x_test,y_test)
plt.scatter(x_test, y_pred)
print(model.named_steps['fitter'].coef_)
plt.show()
```
